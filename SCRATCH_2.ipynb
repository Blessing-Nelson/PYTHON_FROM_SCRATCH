```python
'''TASK 8: Group Names by Their Initial Letter
Goal: Write a program that takes a list of names and groups them into a dictionary where the keys are the first letters of the names and the values are lists of names that start with that letter.

Example Input:
names = ["Alice", "Adam", "Bob", "Charlie", "Cathy"]'''

names = ["Alice", "Adam", "Bob", "Charlie", "Cathy"]

dictionary = {}

for name in names:
    letter = name[0]
    if letter not in dictionary:
        dictionary[letter] = []
    dictionary[letter].append(name)
print(dictionary)
```

```python
'''TASK 9:Advanced - Anagram Groups
Goal: Given a list of words, group them into a dictionary where the keys are sorted versions of the words, and the values are lists of words that are anagrams.

Example Input:

python
Copy code
words = ["listen", "silent", "enlist", "cat", "act", "tac", "hello"]
'''
words = ["listen", "silent", "enlist", "cat", "act", "tac", "hello"]

anagram_groups = {}

for word in words:
    sorted_word = ''.join(sorted(word))     # Sort the letters of the word and join them back into a string

    if sorted_word not in anagram_groups:       # Check if the sorted word is already a key in the dictionary
        anagram_groups[sorted_word] = []        # If not, create an empty list for it

    anagram_groups[sorted_word].append(word)    # Add the original word to the list corresponding to the sorted word


print("Anagram groups:", anagram_groups)
```

```python
'''Task 5: Leap Year Checker
Create a Python program that checks whether a given year is a leap year.
A leap year is divisible by 4, but not divisible by 100 unless it's also divisible by 400.'''

print('Checking leap year')

year = int(input('Enter year: '))
if year % 4 == 0:
    if year % 100 != 0 or year % 400 == 0:
        print(year, 'is a leap year')
    else:
        print(year, 'isn\'t a leap year')
else: 
    print(year, 'isn\'t a leap year')
```
