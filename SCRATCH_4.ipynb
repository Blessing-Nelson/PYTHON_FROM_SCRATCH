```python
'''TASK 2: Encapsulation
Create a class BankAccount with the following features:

Private attribute balance.
Public methods deposit(amount), withdraw(amount), and check_balance().
Prevent withdrawals if the balance is insufficient and print an error message.
'''


class BankAccount:
    def __init__(self, balance):
        self.__balance = balance        #private attribute
    
    def deposit(self, amount):
        '''Function that adds deposited amount to the account balance'''

        print(f'_________Depositing______')
        print(f"\nDeposit amount: ${amount}")
        self.__balance += amount
        print(f'Successful. ${amount} has been deposited')


    def withdraw(self, amount):
        """Function that subtracts amount when withdrawal is made"""
        print(f'_________Withdrawal______')
        print(f"\nWithdraw amount: ${amount}")
        if self.__balance < amount:
            print(f'Insufficient fund.')
        else: 
            self.__balance -= amount
            print(f'Successful withdrawal.')

    def check_balance(self):
        print(f'\n__________Account Balance________')
        print(f'Balance: ${self.__balance}')


account1 = BankAccount(1000)
account1.deposit(500)

account1.withdraw(300)

account1.check_balance()
```

```python
'''TASK 3: Inheritance
Task: Create a parent class Vehicle with a method start() that prints "The vehicle is starting."

Create two child classes:

Car with an overridden start() method that prints "The car is starting."
Bike with its own start() method that prints "The bike is starting."'''




class Vehicle:          #parent class
    def start(self):
        print("The vehicle is starting.")

class Car(Vehicle):         #1st child class
    def start(self):
        print("The car is starting.")
    

class Bike(Vehicle):            #2nd child class
    def start(self):
        print("The bike is starting.")


# Instantiating objects and testing
vehicle = Vehicle()
car = Car()
bike = Bike()

vehicle.start()
car.start()
bike.start()    

```

```python
'''TASK 4: Polymorphism
Task: Create a parent class Animal with a method make_sound(). Create two child classes:

Cat, which implements make_sound() as "Meow."
Dog, which implements make_sound() as "Woof."
Write a function that takes a list of animals and calls their make_sound() method.'''




class Animal:                          #parent class
    def make_sound(self):
        pass

class Cat(Animal):              #1st child class
    def make_sound(self):
        print('Meow')

class Dog(Animal):
    def make_sound(self):
        print('woof')
    
def lists_animals(animals):
    for animal in animals:
        animal.make_sound()

# List of animal objects
animals = [Dog(), Cat()]

#function call
lists_animals(animals)
```

```python
 
'''TASK 5: Real-World Application: Library System
Task: Create a library management system using OOP:

Create a class Library that has:
A list of books available.
Methods to add books, borrow books, and return books.
Borrowed books should not be available for borrowing until they are returned.
Example:

python
Copy code
library = Library(["The Hobbit", "1984", "The Great Gatsby"])
library.display_books()  # Shows available books
library.borrow_book("1984")  # Borrows the book
library.display_books()  # Updates the list
library.return_book("1984")  # Returns the book
Expected Output:

python
Copy code
Available books: The Hobbit, 1984, The Great Gatsby
You've borrowed '1984'
Available books: The Hobbit, The Great Gatsby
'1984' has been returned.'''


class Library:
    def __init__(self, books):
        # Instance variable for books available in the library
        self.available_books = books
        self.borrowed_books = []  # Track borrowed books separately

    def add_books(self, title):
        print(f"\nBook to add: {title}")
        if title in self.available_books:
            print(f"'{title}' is already in the library!")
        else:
            self.available_books.append(title)
            print(f"'{title}' has been added to the library.")

    def borrow_book(self, title):
        print(f"\nAttempting to borrow: '{title}'")
        if title not in self.available_books:
            print(f"'{title}' is not available for borrowing.")
        else:
            self.available_books.remove(title)
            self.borrowed_books.append(title)
            print(f"You've borrowed '{title}'.")

    def return_book(self, title):
        print(f"\nAttempting to return: '{title}'")
        if title not in self.borrowed_books:
            print(f"'{title}' was not borrowed.")
        else:
            self.borrowed_books.remove(title)
            self.available_books.append(title)
            print(f"'{title}' has been returned to the library.")

    def display_books(self):
        print("\nAvailable books in the library:")
        if not self.available_books:
            print("No books available!")
        else:
            print(", ".join(self.available_books))


# Initialize the library with a list of books
library = Library(["The Hobbit", "1984", "The Great Gatsby"])

# Display the books in the library
library.display_books()

# Borrow a book
library.borrow_book("1984")

# Check available books again
library.display_books()

# Attempt to borrow a book that's already borrowed
library.borrow_book("1984")

# Return the borrowed book
library.return_book("1984")

# Display available books after the return
library.display_books()

# Add a new book
library.add_books("To Kill a Mockingbird")

# Add a duplicate book
library.add_books("1984")

# Final state of the library
library.display_books()

```

```python
'''Task 1: Temperature Checker
Create a program that checks if the entered temperature is within the range of -100°C to 60°C. If not, raise a TemperatureOutOfRangeError.

Define the custom exception TemperatureOutOfRangeError.
Use a function check_temperature(temp) to validate the temperature.
Catch and handle the custom exception if raised.'''



class TemperatureOutOfRangeError (Exception):
    """Custom exception for temperature out of range."""
    
    def __init__(self, message = 'Temperature must be greater than -100 and less than 60.'):
        self.message = message
        super().__init__(self.message)

def check_temperature(temp):
    if not (-100 <= temp <= 60):  # Correct logical comparison
        raise TemperatureOutOfRangeError()
    else:
        print(f'Inputted temperature is: {temp}') 




try:
   user = float(input("Enter temperature (°C): "))  
   check_temperature(user)
except TemperatureOutOfRangeError as e:
    print(e)
except ValueError:
    print('Invalid input. Please enter a numeric value.')
finally:
    print('Thank you for using python temperature checker.')
```

```python
'''Task 2: Student Grading System
Write a program where:

A validate_grade(grade) function checks if a grade is between 0 and 100.
Raise a custom exception InvalidGradeError if the grade is out of range.
Handle the exception and prompt the user to re-enter a valid grade.
'''

class InvalidGradeError(Exception):

    '''Custom expression for grades.'''

    def __init__(self, message='Grade must be between 0 and 100'):
        self.message = message
        super().__init__(message)

def validate_grade(grade):

    '''Validates if the grade is between 0 and'''
    
    if not(0 <= grade <= 100):
        raise InvalidGradeError
    else: 
        print(f'Grade {grade} is valid.')

while True:
    try:
    
        user = float(input('Enter your grade: '))
        validate_grade(user)
        break
    except InvalidGradeError as e:
        print(e)
    except ValueError:
        print('Invalid error. Input a numeric value.')
```

```python
'''TASK 4: Password Validation
Create a program that validates passwords based on the following rules:

Must be at least 8 characters.
Must contain at least one uppercase letter, one lowercase letter, and one digit.
If the password is invalid, raise a PasswordValidationError.
'''

class PasswordValidationError(Exception):
    def __init__(self, message= "Invalid password."):
        self.message = message
        super().__init__(self.message)


def password_checker(password):
    if len(password) < 8:                                                           # Checks  length requirement
        raise PasswordValidationError('Password must be at least 8 characters')
    
    if not any(char.isupper() for char in password):                                # Checks for at least one uppercase letter
        raise PasswordValidationError('Password must contain at least one uppercase letter')
    
    if not any(char.islower()for char in password):                                  # Checks for at least one lowercase letter
        raise PasswordValidationError( 'Password must contain at least one lowercase letter')
    
    if not any(char.isdigit() for char in password):                                 # Checks for at least one digit
        raise PasswordValidationError( "Input must contain at least one digit.")
    
    return 'Password is valid.'
    
while True:                                                                           #Code runs untl the correct password in inputted
        try:
            user = input('Enter password: ')
            password_checker(user)
            print(f'Password is {user}')
            break

        except PasswordValidationError as e:
            print(e)
```
